```python
import pandas as pd
import requests
from io import StringIO
from google.colab import files
```

```python
url = "https://www.howstat.com/Cricket/Statistics/Series/SeriesAnalysis.asp?SeriesCode=0863&Scope=All"
```

```python
response = requests.get(url)
html_read = StringIO(response.text)
table  = pd.read_html(html_read)
```

```python
df = table[3]
```

```python
df
```

```python
df = df.iloc[:, :-3]
df
```

```python
df.to_csv("battingstats.csv",index= "False")
files.download("battingstats.csv")
```
